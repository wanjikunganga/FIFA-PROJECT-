# 6) Implementing the solution
At this point, we are going to develop a model and train it to make predictions.

The first step is data pre processing where we will split into training and testing set: Using Stratified Split since our class is imbalanced to ensure we have equal propotions in both train and test sets


## a) Data Preparation

```python
# dropping date column
df1.drop(['date'], inplace = True, axis = 1)
```

```python
# a function to get all columns of object type
obj_list = df1.select_dtypes(include = "object").columns
print (obj_list)
```

```python
#Label Encoding for object to numeric conversion
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()

for feat in obj_list:
    df1[feat] = le.fit_transform(df1[feat])

print (df1.info())
```

```python
# Convert all columns to int dtype.
df1 = df1.astype('int')
```

## b) Train and Test Split



```python
# Split the independent and dependent variables
# Home team result will be our Y variable.
# 
y = df1['home_team_result']
```

```python
#Converting y into categorical type
from keras.utils.np_utils import to_categorical
y_cat = to_categorical(y)
```

```python
# assigning our independent variables
X = df1.drop(['home_team_result'], axis = 1)
```

```python
# Train using 80% of the data.
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y_cat, test_size=0.2, random_state=0, stratify = y)
```

```python
# previewing our target variable to see the propotions in the train set
```

```python
y_train.value_counts(normalize = True) * 100
```

```python
# previewing our target variable to see the propotions in the test set
y_test.value_counts(normalize = True) * 100
```

## c) Data Preprocessing

```python
# data normalization with sklearn
from sklearn.preprocessing import MinMaxScaler

# fit scaler on training data
norm = MinMaxScaler().fit(X_train)

# transform training data
X_train_norm = norm.transform(X_train)

# transform testing data
X_test_norm = norm.transform(X_test)
```

## c) Importing dependencies

```python
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import models
from tensorflow.keras.models import Sequential, load_model 
from tensorflow.keras.layers import Dense
```

## d) Build and Compile Model


*  Dense 64 is a fully connected  layer with 64 hidden units

* The expected input data shape is the number of columns in the X train set

* The final layer with 3 units same as the number of classes



```python
model = keras.Sequential()
model.add(Dense(units = 60, activation = 'relu', input_dim = len(X_train.columns)))
model.add(Dense(units =  15, activation = 'relu'))
model.add(Dense(units = 3, activation = 'softmax'))
opt = keras.optimizers.Adam(learning_rate=0.001)
model.compile(loss = 'categorical_crossentropy', optimizer = opt, metrics = ['accuracy','Precision', 'Recall'])
model.summary()
```

## e) Fit, Predict, Evaluate


##### i) Fitting the model

```python
# fitting the model
tf.keras.callbacks.History()
history = model.fit(X_train_norm, y_train, epochs = 60, batch_size = 33)
```

```python
# plotting the loss against epoch size to determine epoch size to maintain
import matplotlib.pyplot as plt
from matplotlib import rcParams

rcParams['figure.figsize'] = (18,8)
rcParams['axes.spines.top'] = False
rcParams['axes.spines.right'] = False

plt.plot(np.arange(1,61),history.history['loss'], label = 'Loss')
plt.title('Loss Against Epoch Size', size = 20)
plt.xlabel('Epoch', size = 13)
plt.ylabel('Loss', size = 13);
```

**Observation**

Loss reduces as we train the model, with occasional spikes with a sign of plateau at we approch 60. We wont train our model further to avoid overfitting



##### ii) Making Predictions

```python
# using the test data to make predictions
predict_x=model.predict(X_test_norm) 
y_pred_classes=np.argmax(predict_x,axis=1)

y_test_class = predictions = np.argmax(y_test, axis=1)

y_pred = model.predict(X_test)
```

##### iii) Evaluating the model

```python
# importing relevant libraries
from sklearn.metrics import confusion_matrix
from sklearn.metrics import accuracy_score
from sklearn.metrics import classification_report, confusion_matrix
```

```python
#print confusion matrix
confusion_matrix(y_test_class, y_pred_classes)

```

4753 out of 4785 records were predicted correctly.

```python
#print classification report
classification_report(y_test_class, y_pred_classes)
print(classification_report(y_test_class, y_pred_classes))
```

Our precision score, recall and F1 score is 100%. Our accuracy is 99%.
